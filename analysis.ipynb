#  Size of population, vital statistics and migration in 1950-2023

Import of necessary libraries

```python
import pandas as pd
import numpy as np
```

 Upload property file with data

```python
dataset = pd.read_csv("poland-population.csv")
dataset.info()
```

```python
for col in dataset.columns:
    if dataset[col].dtype == "object":
        print(f"Column {col} has {dataset[col].nunique()} unique values")
```

```python
dataset.head()
```

```python
for col in dataset.columns[1:15]:
    dataset = dataset.rename(columns={col:col+" in thousands"})
```

```python
dataset.columns
```

```python
for col in dataset.columns:
    if ".1" in col:
        dataset = dataset.rename(columns={col:col[:-2]})
```

```python
dataset.columns
```

```python
for col in ['Marriages', 'Divorces','Live births', 'Deaths', 'Natural increase', 'Total net migration']:
    dataset = dataset.rename(columns={col:col+" per 1000 population"})
```

```python
dataset.columns
```

```python
dataset = dataset.drop(columns=["Internal migration net in thousands"])
```

```python
for col in dataset.columns:
    if dataset[col].dtype == "object":
        dataset[col] = dataset[col].str.replace('c','')
        dataset[col] = dataset[col].str.replace(',','.')
        dataset[col] = dataset[col].astype("float")
```

```python
for col in dataset.columns:
    if (dataset[col].dtype == "float" or dataset[col].dtype == "int") and ("in thousands" in col):
        dataset[col] = dataset[col]*1000
        dataset = dataset.rename(columns={col:col[:-13]})
```

```python
dataset.columns
```

```python
dataset.head()
```

```python
dataset[dataset["Years"]==2015]
```

```python
for col in dataset.columns:
    if dataset[col].dtype == "float":
        dataset[col] = dataset[col].astype("int")
```

```python
dataset.head()
```

```python
thousands_df = dataset.iloc[::,:14:]
per_population_df = dataset.iloc[:,np.r_[0, 14:dataset.shape[1]]]
```

```python
thousands_df.to_csv("thousands_population.csv")
per_population_df.to_csv("per_population.csv")
```
